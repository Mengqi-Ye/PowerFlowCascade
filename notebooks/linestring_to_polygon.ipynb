```python
import os
import sys
from pathlib import Path

import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString, mapping

parent_dir = Path().resolve().parent
sys.path.insert(0, str(parent_dir))
```

```python
subs_gdf = gpd.read_file("../data/osm/vietnam-latest_substation_supplyPolygonID.gpkg")
subs_gdf = subs_gdf.to_crs(epsg=32648)
subs_gdf_line = subs_gdf[subs_gdf.geometry.type == 'LineString']
```

```python
# https://stackoverflow.com/questions/2964751/how-to-convert-a-geos-multilinestring-to-polygon
def linestring_to_polygon(gdf):
    gdf['geometry'] = [Polygon(mapping(x)['coordinates']) for x in gdf.geometry]
    return gdf
```

```python
subs_gdf_line_to_polygon = linestring_to_polygon(subs_gdf_line)
```

```python
subs_gdf_line_to_polygon.to_file('../outputs/osm_subs_gdf_linestring_to_polygon.gpkg', driver='GPKG',espsg='32648')
```
